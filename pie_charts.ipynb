```python
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5]
```

```python
sleeping = [7,8,6,11,7]
eating   = [2,3,4,3,2]
working  = [7,8,7,2,2]
playing  = [8,5,7,8,13]
```

```python
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','k']

plt.pie(slices, labels=activities, colors=cols)
plt.xlabel('x')
plt.ylabel('y')
plt.title('intresting graph')
plt.legend()
plt.show()
```

```python
plt.pie(slices, labels=activities, colors=cols,startangle=90)
```

```python
plt.pie(slices, labels=activities, colors=cols,startangle=90, shadow=True)
```

```python
plt.pie(slices, labels=activities, colors=cols,startangle=90, shadow=True, explode=(0,0.1,0,0))
```

```python
plt.pie(slices, 
        labels=activities, 
        colors=cols,
        startangle=90, 
        shadow=True, 
        explode=(0,0.1,0,0),
        autopct='%1.1f%%')
```
