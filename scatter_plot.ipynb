```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,5,2,5,6,3]

plt.scatter(x,y, label='skitscat', color='b')

plt.xlabel('x')
plt.ylabel('y')
plt.title('intresting graph')
plt.legend()
plt.show()
```

```python
plt.scatter(x,y, label='skitscat', color='b',marker="o")
```

```python
plt.scatter(x,y, label='skitscat', color='b',marker="o",s=100)
```

```python
plt.scatter(x,y, label='skitscat', color='b',marker="*",s=100)
```
